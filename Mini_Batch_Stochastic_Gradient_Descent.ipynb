```python
def create_mini_batches(X, y, batch_size):
    mini_batches = []
    data = np.hstack((X, y))
    np.random.shuffle(data)
    n_minibatches = data.shape[0] // batch_size
    i = 0
 
    for i in range(n_minibatches + 1):
        mini_batch = data[i * batch_size:(i + 1)*batch_size, :]
        X_mini = mini_batch[:, :-1]
        Y_mini = mini_batch[:, -1].reshape((-1, 1))
        mini_batches.append((X_mini, Y_mini))
    if data.shape[0] % batch_size != 0:
        mini_batch = data[i * batch_size:data.shape[0]]
        X_mini = mini_batch[:, :-1]
        Y_mini = mini_batch[:, -1].reshape((-1, 1))
        mini_batches.append((X_mini, Y_mini))
    return mini_batches
```

```python
def SGD(data, learning_rate=0.01, n_epochs=1000, k=10, batch_size=40):
    w = np.random.randn(1,data.shape[1]-1)
    b = 0
    epoch=1
    
    while epoch <= n_epochs:
        temp = data.sample(k)
        y = temp.iloc[:,4:].values.tolist()
        x = np.array(temp.drop("class",axis=1))
        L_w = w
        L_b = b
        loss = 0
        y_pred = []
        
        for i in range(k):
            mini_batches = create_mini_batches(x, y, batch_size)
            for mini_batch in mini_batches:
                x, y = mini_batch
                L_w = (-2/k * x[i]) * (y[i] - np.dot(x[i],w.T) - b)
                L_b = (-2/k) * (y[i] - np.dot(x[i],w.T) - b)
                w = w - learning_rate * L_w
                b = b - learning_rate * L_b
            y_predicted = np.dot(x[i],w.T)
            y_pred.append(y_predicted)
            
        loss = mean_squared_error(y_pred, y)
        print("Epoch: %d, Loss: %.3f" %(epoch, loss))
        epoch += 1
        learning_rate = learning_rate/1.02
        
    return w,b
```

```python
def predict(x,w,b):
    y_pred = []
    for i in range(len(x)):
        y = np.asscalar(np.dot(w,x[i])+b)
        y_pred.append(y)
    return np.array(y_pred)
```
