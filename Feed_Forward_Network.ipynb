```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from keras.layers import Dense
from keras.models import Sequential
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Dry_Bean_Dataset.csv')
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,0:16]
X 
```

```python
Y = data.iloc[:,16:]
Y
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
dummy_Y = np_utils.to_categorical(encoded_Y)
```

```python
dummy_Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,dummy_Y, test_size = 20, random_state = 1)
```

```python
model = Sequential()
model.add(Dense(9, input_dim = 16, activation = 'relu'))
model.add(Dense(7, activation = 'relu'))
model.add(Dense(7, activation = 'relu'))
model.add(Dense(7, activation = 'relu'))
model.add(Dense(7, activation = 'sigmoid'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs = 200, batch_size = 128) 
```

```python
loss, accuracy = model.evaluate(X_train, Y_train,batch_size = 128)
```

```python
print("Accuracy: %.2f%% Loss: %.2f%%" % (accuracy*100, loss*100))
```

```python
Actual = np.argmax(model.predict(X_test),axis=1)
```

```python
for i in range(len(Y_test)):
  print("X = %s, Predicted = %s" % (Y_test[i], Actual[i]))
```
