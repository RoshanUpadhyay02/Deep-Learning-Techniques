```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import MinMaxScaler
from keras.optimizers import SGD
```

```python
data = pd.read_csv("data_banknote_authentication.txt",sep = ',',header = None)
data
```

```python
data.rename(columns = {0: "variance_of_Wavelet_Transformed_image",
                       1: "skewness_of_Wavelet_Transformed_image", 
                       2: "curtosis_of_Wavelet_Transformed_image", 
                       3: "entropy_of_image", 
                       4: "class"}, inplace = True)
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
X = data.iloc[:,:4]
X
```

```python
Y = data.iloc[:,4:]
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 20 , random_state = 1)
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
error = mean_absolute_error(y_pred,Y_test)
print(error)
```

```python
score = r2_score(y_pred,Y_test)
print(score)
```

```python
model = SGDRegressor(loss='squared_error',penalty='l2', alpha=0.0001)
model.fit(X_train,Y_train.values.ravel())
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
error = mean_absolute_error(y_pred,Y_test)
print(error)
```

```python
score = r2_score(y_pred,Y_test)
print(score)
```
