```python
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
def encode(Y):
  encoder = LabelEncoder()
  encoder.fit(Y)
  encoded_Y = encoder.transform(Y)
  dummy_Y = np_utils.to_categorical(encoded_Y)
  return dummy_Y
```

```python
data['gender'] = encode(data['gender'])
data['ever_married'] = encode(data['ever_married'])
data['work_type'] = encode(data['work_type'])
data['Residence_type'] = encode(data['Residence_type'])
data['smoking_status'] = encode(data['smoking_status'])
```

```python
import numpy as np
w = np.zeros((4909 ,10))
w
```

```python
for i in range(4909):
    for j in range(10):
        a = round(random.uniform(0, 1), 2)
        w[i][j] = a
w.astype(float)
w
```

```python
b = round(random.uniform(0, 1), 2)
b
```

```python
def sigmoid_function(x):
    return 1/(1 + np.exp(-x))
```

```python
def sigmoid_deriv(x):
    return x * (1 - x)
```

```python
def error(out,i):
    return output[i] - out
```

```python
errors = np.zeros((4909 ,10)).astype(float)
errors
```

```python
y_pred = []
for i in range(4909 ):
    sum = 0
    out = 0
    a = 0
    for j in range(10):
        sum += input_[i][j]*w[i][j] + b
    out = sigmoid_function(sum)
    a = -1*error(out,i)
    q = out*(1-out)
    p = input_[i][j]
    
    if a != 0:
        errors[i][j] = a
    while a != 0:
        for j in range(8):
            sum += input_[i][j]*w[i][j] + b
        out = sigmoid_function(sum)
        a = error(out,i)
        print("error = %.2f output = %.2f" %(a,out))
        if a != 0:
            errors[i][j] = a 
        for j in range(8):
            w[i][j] = w[i][j] + a*output[i]
        b = b + a
        print("updated_weight = %.2f updated_bias = %.2f" %(w[i][j],b))
    y_pred.append(out)
    print("Brain Stroke({}) = {}".format(input_[i],out))
    print()
```
